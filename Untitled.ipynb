```python
import pandas as pd
import re
```

```python
f_rashi=open('_chat_rashi.txt',encoding='utf-8')
data_rashi=f_rashi.read()
data_rashi
```

```python
print(data_rashi)
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[a-zA-Z]+\s-\s'
# pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[a-zA-Z]'
```

```python
messages=re.split(pattern,data_rashi)[1:]
messages
```

```python
# dates=re.findall(pattern, data_rashi)
# dates
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s'
dates=re.findall(pattern,data_rashi)
dates
```

```python
df=pd.DataFrame({'user_messages':messages,'message_dates':dates})
df.head(5)
```

```python
# df['message_dates'] = pd.to_datetime(df['message_dates'], format='%d/%m/%Y, %H:%M ')
df['message_dates'] = pd.to_datetime(df['message_dates'])

df
```

```python
df.rename(columns={'message_dates':'dates'},inplace=True)
df.head()
```

```python
users=[]
messages=[]
for message in df['user_messages']:
    entry=re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group notification')
        messages.append(entry[0])
df['user']=users
df['messages']=messages
df.drop(columns=['user_messages'],inplace=True)
```

```python
period=[]
for hour in df[['day_name','hour']]['hour']:
    if hour==23:
        period.append(str(hour)+"-"+"00")
    elif hour==0:
        period.append(str("00")+"-"+str(hour+1))
    else:
        period.append(str(hour)+"-"+str(hour+1))
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour + 1))
    else:
        period.append(str(hour) + "-" + str(hour + 1))

df['period'] = period
```

```python
df['period']=period
```

```python
df['period'].value_counts()
```

```python
import seaborn as sns
plt.figure(figsize=(18,10))
sns.heatmap(df.pivot_table(index='day_name',columns='period',values='messages',aggfunc='count').fillna(0))
plt.xticks(rotation='horizontal')
plt.show()
```
